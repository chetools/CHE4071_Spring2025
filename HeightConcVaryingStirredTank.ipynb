```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g=9.81 #m/s2
Ac = 5. #m2/s
Aout = 0.1 #m2/s

qin0 = 0.2 #m3/s
h0 = ((qin0/Aout)**2)/(2*g)

Cain0 = 100. #mol/m3
Ca0 = Cain0
tend = 100. #s
```

```python
def rhs(t, vec):
    Ca, h = vec
    Cain = Cain0 + 0.1*Cain0*np.sin(t)
    qin = 0.9 * qin0

    qout = Aout*np.sqrt(2*g*h)
    dhdt = (qin - qout)/Ac
    dCadt = (Cain * qin - qout*Ca - Ca*Ac*dhdt)/(Ac*h)

    return [dCadt, dhdt]
```

```python
res=solve_ivp(rhs , (0,tend), [Ca0, h0], method='Radau', dense_output=True)
```

```python
tplot = np.linspace(0,tend,200)
Caplot, hplot = res.sol(tplot)
```

```python
fig=make_subplots(rows=1, cols=2)
fig.add_scatter(x=tplot, y= Caplot, row=1, col=1, name='Ca')
fig.add_scatter(x=tplot, y=hplot, row=1, col=2, name='h')
fig.update_layout(width=800, height=400, template='plotly_dark')
```
