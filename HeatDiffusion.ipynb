```python
import numpy as np
from scipy.linalg import solve_banded
from plotly.subplots import make_subplots
```

```python
r=0.50016
N=100
T = np.full(N, 20.)
T[0] = 50.
T[-1]=80.

fig = make_subplots()
pos = np.arange(N)
fig.add_scatter(x=pos, y=T, mode='lines', name=f'0')
for i in range(50000):
    T[1:-1] = T[1:-1] + r*(T[2:] - 2*T[1:-1] + T[:-2])
    if i % 1000 ==0:
        fig.add_scatter(x=pos, y = T, mode='lines',name=f'{i}')

fig
```

```python
#Crank-Nicolson
r=1  #r = alpha*(dt)/(2*(dx^2))
N=100
T = np.full(N, 20.)  #Just the unknown temperatures in between
Tleft= 50
Tright = 80.
sparse_A = np.c_[np.full(N, -r),np.full(N, 1+2*r),np.full(N, -r)].T
fig = make_subplots()
pos = np.arange(N+2)
fig.add_scatter(x=pos, y=np.r_[Tleft, T, Tright], mode='lines', name=f'0')
for i in range(1,5000):
    b = (1-2*r)*T
    b[1:-1]+= r*T[2:]
    b[1:-1]+=r*T[:-2]
    b[0]+= r*T[1] + 2*r*Tleft
    b[-1]+=r*T[-2] + 2*r*Tright
    T=solve_banded((1,1), sparse_A, b)
    if i % 100 ==0:
        fig.add_scatter(x=pos, y =np.r_[Tleft, T, Tright], mode='lines',name=f'{i}')

fig
```
