```python
import numpy as np
from plotly.subplots import make_subplots
```

```python
r=0.50017
N=100
T = np.full(N, 20.)
T[0] = 50.
T[-1]=80.

fig = make_subplots()
pos = np.arange(N)
fig.add_scatter(x=pos, y=T, mode='lines', name=f'0')
for i in range(50000):
    T[1:-1] = T[1:-1] + r*(T[2:] - 2*T[1:-1] + T[:-2])
    if i % 1000 ==0:
        fig.add_scatter(x=pos, y = T, mode='lines',name=f'{i}')

fig
```
