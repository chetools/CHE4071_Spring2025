```python
Gp = (2/(10*s+1) - 1/(2*s+1)); Gp
```

```python
Kc = -1.1
Gcl = Kc*Gp/(1+Kc*Gp)
```

```python
t, y = sim(Gcl, lambda t: 1., N=1000, dt=2)
```

```python
fig = make_subplots()
fig.add_scatter(x=t, y=y, mode='lines')
fig.update_layout(width=600, height=400, template='plotly_dark')
```
