```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
A = 2.
alpha = 0.2
qin0 = 0.5
qin = 0.55
h0 = (qin0/alpha)**2

a = qin-qin0
K= 2*np.sqrt(h0)/alpha
tau = K*A
```

```python
def rhs(t, v):
    h=v[0]
    dh = (qin - alpha*np.sqrt(h))/A
    return [dh]
```

```python
tend= 500
res=solve_ivp(rhs, (0,tend), [h0], method='Radau', dense_output=True)
tplot = np.linspace(0,tend, 100)
h=res.sol(tplot)[0]
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot, y=h, name='exact')
fig.add_scatter(x=tplot, y=h0 + a*K*(1-np.exp(-tplot/tau)), name='linearized')
fig.update_layout(width=600, height=400, template='plotly_dark')
```
