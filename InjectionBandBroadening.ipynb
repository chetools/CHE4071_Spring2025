```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.integrate import solve_ivp
from scipy.special import expit
```

```python
totV = 1. #L
N = 20
q = 0.1 #L/s
V = totV/N
ca0 = 0.1
c0 = np.full(N, ca0)
```

```python
def cin(t):
    return 0.1 + 0.4*expit(50*(t-1)) - 0.4*expit(50*(t-2))
```

```python
tplot= np.linspace(0,10, 500)
fig= make_subplots()
fig.add_scatter(x=tplot, y = cin(tplot), mode='lines')
fig.update_layout(width=800, height=400, template='plotly_dark')
```

```python
def rhs(t, c):
    dc0 = q/V*(cin(t) - c[0])
    dc1_n = q/V*(c[:-1]- c[1:])
    return np.r_[dc0, dc1_n]

```

```python
tend = 20
tplot= np.linspace(0,tend, 200)
res=solve_ivp(rhs, (0, tend), c0, method='Radau', dense_output=True)
cs = res.sol(tplot)
```

```python
fig2= make_subplots()
for i,c in enumerate(cs):
    fig2.add_scatter(x=tplot, y = c, mode='lines', name=f'tank {i}')
fig2.update_layout(width=800, height=400, template='plotly_dark')
```

```python
fig3= make_subplots()
xplot = np.linspace(0,1, N)
for i in range(0,200,10):
     fig3.add_scatter(x=xplot, y = cs[:,i], mode='lines', name=f'time {i}')
fig3.add_scatter(x=xplot, y = cs[:,-1], mode='lines', name=f'time 200')
fig3.update_layout(width=800, height=400, template='plotly_dark')
```
